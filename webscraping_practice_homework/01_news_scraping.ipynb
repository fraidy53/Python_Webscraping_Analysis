```python
import requests
from bs4 import BeautifulSoup

url = 'https://news.daum.net/economy'

response = requests.get(url)
response.encoding = 'utf-8'

print(response.url)
print(type(response))
print(response.status_code)

soup = BeautifulSoup(response.text, 'html.parser')

news_list = soup.select('ul.list_newsheadline2 li')
print(type(news_list), len(news_list))

for li_tag in news_list:
    a_tag = li_tag.find('a')
    link = a_tag['href']
    
    strong_tag = li_tag.select_one('div.cont_thumb strong.tit_txt')
    title = strong_tag.text.strip() #.text : 태그 안 글자만 가져오기, .strip() : 앞뒤 공백 제거
    
    print(link)
    print(title)
```

```python
import requests
from bs4 import BeautifulSoup

section_dict = {
    '기후/환경':'climate',
    '사회':'society',
    '경제':'economy',
    '정치':'politics',
    '국제':'world',
    '문화':'culture',
    '생활':'life',
    'IT/과학':'tech',
    '인물':'people'
}

def print_news(section_name):
    # 1. 한글 섹션 → 영문 섹션
    eng_section = section_dict[section_name]
    
    # 2. URL 생성
    url = f'https://news.daum.net/{eng_section}'
    
    print(f'======> {url} {section_name} 뉴스 <======')
    
    # 3. 웹 요청
    response = requests.get(url)
    response.encoding = 'utf-8'
    
    # 4. HTML 파싱
    soup = BeautifulSoup(response.text, 'html.parser')
    news_list = soup.select('ul.list_newsheadline2 li')
    
    # 5. 기사 출력
    for li_tag in news_list:
        a_tag = li_tag.find('a')
        link = a_tag['href']
        
        strong_tag = li_tag.select_one('div.cont_thumb strong.tit_txt')
        title = strong_tag.text.strip()
        
        print(link)
        print(title)

# 함수 호출
print_news('경제')
print_news('사회')
```
